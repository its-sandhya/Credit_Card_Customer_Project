```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
from urllib.parse import quote_plus
```

```python
# # Step 1: Read CSV files

path = 'C:/Users/lenovo/OneDrive/Desktop/Machine Learning/PROJECTCS/2 BI Credit_Card_Financial_Dashboard/data csv files/'

credit_card_df = pd.read_csv(path + 'credit_card.csv')
customer_df = pd.read_csv(path + 'customer.csv')
cust_add_df = pd.read_csv(path + 'cust_add.csv')
cc_add_df = pd.read_csv(path + 'cc_add.csv')
```

```python
# Step 2: Connect to MySQL and create database
conn = pymysql.connect(host="localhost", user="root", password="Sandhya@25", port=3306)
cur = conn.cursor()
cur.execute("CREATE DATABASE IF NOT EXISTS ccdb;")
conn.close()
print("✅ Database 'ccdb' created or already exists.")
```

```python
# Step 3: Create SQLAlchemy engine for 'ccdb'

username = "root"
password = quote_plus("Sandhya@25")
host = "localhost"
port = 3306
database = "ccdb"
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")
```

```python
# Step 4: Import CSVs into MySQL

credit_card_df.to_sql("credit_card", engine, if_exists="replace", index=False)
customer_df.to_sql("customer", engine, if_exists="replace", index=False)
cust_add_df.to_sql("cust_add", engine, if_exists="replace", index=False)
cc_add_df.to_sql("cc_add", engine, if_exists="replace", index=False)

print("✅ All 4 CSV files imported successfully into 'ccdb'.")
```
